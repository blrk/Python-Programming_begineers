### Arithmetic and Operator Overloading
* For a built-in type such as int or float , each arithmetic operator corresponds to a special method name. 
* You will see many of these methods by entering dir(int) or dir(str) at a shell prompt, and they are listed below. 
* The object on which the method is called corresponds to the left operand,* 

Screenshot 2022-03-30 at 9.06.50 AM.png

```python
class Number:
  def __init__(self, x):
    self.x = x
  def getNumber(self):
    print("Value of x:", self.x)
```

```python
n1 = Number(10)
n2 = Number(20)
```

```python
n1.getNumber()
n2.getNumber()
```

```python
a = 10 
b = 20
a + b
```

```python
n1 + n2
```

```python
class Number:
  def __init__(self, x):
    self.x = x
  def getNumber(self):
    print("Value of x:", self.x)
  def __add__(self, obj2): # operator overlaoding function
    return self.x + obj2.x
```

```python
n1 = Number(10)
n2 = Number(20)
```

```python
n1 + n2
```

### Comparison Methods
* You can compare integers and floating-point numbers using the operators == , ! = , < , > , <= ,and >= . 
* When the Python interpreter encounters one of these operators, it uses a corresponding method defined in the float or int class. 
* Each of these methods expects two arguments. The first argument, self , represents the operand to the left of the operator, and
the second argument represents the other operand. 
* Following table lists the comparison operators and the corresponding methods.

Screenshot 2022-03-30 at 9.08.14 AM.png

```python
class Number:
  def __init__(self, x):
    self.x = x
  def getNumber(self):
    print("Value of x:", self.x)
```

```python
n1 = Number(20)
n2 = Number(5)
```

```python
n1 > n2
```

```python
x, y = 10, 5
x > y
```

```python
class Number:
  def __init__(self, x):
    self.x = x
  def getNumber(self):
    print("Value of x:", self.x)
  def __gt__(self, obj2): # implmementation of > operator overloading fun
    if self.x > obj2.x:
      return True
    else:
      return False
```

```python
n1 = Number(20)
n2 = Number(5)
```

```python
n1 > n2
```

```python
n3 = Number(2)
n4 = Number(5)
```

```python
n3 > n4
```
