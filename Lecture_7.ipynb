```python
# Guessing number game

import random
small = int(input("Enter the smaller number:"))
large = int(input("Enter the larger number:"))
myNum = random.randint(small, large)
count = 0
while True:
  count +=1
  userNum = int(input("Enter your guess:"))
  if userNum < myNum:
    print("Too small")
  elif userNum > myNum:
    print("Too Large")
  else:
    print("wow! you got it at round", count)
    break
```
